```python
# Your code here
import re

#ans a)
sentence1 = str(input("Enter 1st sentence: "))
sentence2 = str(input("Enter 2nd sentence: "))
print("\n")

#ans b)
sent_lis1 = re.findall(r"\w+|[^\w\s]",sentence1)
sent_lis2 = re.findall(r"\w+|[^\w\s]",sentence2)
# full_sent = sent_lis1 + sent_lis2
# print(full_sent)
print(sent_lis1)
print(sent_lis2)
print("\n")

#ans c)
similar_words = list(set(sent_lis1) & set(sent_lis2))
# Filter out non-alphabetic words using a for loop
similar_words_cleaned = []
for word in similar_words:
    if word.isalpha():
        similar_words_cleaned.append(word)

print("Words that occur in both the sentences:", similar_words_cleaned)
print()

#and d)
sant_sent1 = re.sub(r"[^\w\s]", "", sentence1)
sant_sent2 = re.sub(r"[^\w\s]", "", sentence2)
print("Sanitized sentence 1: ", sant_sent1)
print("Sanitized sentence 2: ", sant_sent2)
print("\n")

#ans e)
vowel_count_1 = {"a":0,"e":0,"i":0,"o":0,"u":0}
for char in sentence1.lower():
  if char in vowel_count_1.keys():
    vowel_count_1[char] += 1
print(f"Count of vowels in the sentence 1 is: {vowel_count_1}")

vowel_count_2 = {"a":0,"e":0,"i":0,"o":0,"u":0}
for char in sentence2.lower():
  if char in vowel_count_2.keys():
    vowel_count_2[char] += 1
print(f"Count of vowels in the sentence 2 is: {vowel_count_2}")
```

Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.


#prompt1 :-create me a code to Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options.

```python
import pandas as pd
import os

# File path for the inventory CSV file (local)
FILE_PATH = "/Users/patial31/Desktop.csv"

def load_inventory(file_path):
    """Loads the inventory from a CSV file, or creates a new one if it doesn't exist."""
    if os.path.exists(file_path):
        return pd.read_csv(file_path)
    else:
        df = pd.DataFrame(columns=["name", "category", "quantity", "price", "threshold"])
        df.to_csv(file_path, index=False)
        return df

def save_inventory(df, file_path):
    """Saves the inventory to a CSV file."""
    df.to_csv(file_path, index=False)

def add_product(df):
    """Adds a new product to the inventory or updates the quantity of an existing product."""
    name = input("Enter Product Name: ")

    if name in df["name"].values:
        print("This product already exists.")
        action = input("Do you want to add or delete quantity? (add/delete): ").strip().lower()

        if action == "add":
            quantity_to_add = int(input("Enter quantity to add: "))
            df.loc[df["name"] == name, "quantity"] += quantity_to_add
            print(f"Added {quantity_to_add} to {name}. New quantity: {df.loc[df['name'] == name, 'quantity'].values[0]}")
        elif action == "delete":
            quantity_to_delete = int(input("Enter quantity to delete: "))
            current_quantity = df.loc[df["name"] == name, "quantity"].values[0]
            if current_quantity >= quantity_to_delete:
                df.loc[df["name"] == name, "quantity"] -= quantity_to_delete
                print(f"Deleted {quantity_to_delete} from {name}. New quantity: {df.loc[df['name'] == name, 'quantity'].values[0]}")
            else:
                print("Cannot delete more than the available quantity.")
        else:
            print("Invalid action. Please choose 'add' or 'delete'.")
    else:
        category = input("Enter Product Category: ")
        quantity = int(input("Enter Quantity: "))
        price = float(input("Enter Price: "))
        threshold = int(input("Enter Threshold: "))

        new_product = pd.DataFrame([[name, category, quantity, price, threshold]], columns=df.columns)
        df = pd.concat([df, new_product], ignore_index=True)
        print("Product added successfully.")

    save_inventory(df, FILE_PATH)
    return df

def view_products(df):
    """Displays all products in the inventory."""
    print(df)

def update_product(df):
    """Updates the details of an existing product."""
    name = input("Enter Product Name to update: ")

    if name in df["name"].values:
        index = df[df["name"] == name].index[0]
        print("Existing Details:\n", df.loc[index])

        df.at[index, "name"] = input("Enter New Name: ")
        df.at[index, "category"] = input("Enter New Category: ")
        df.at[index, "quantity"] = int(input("Enter New Quantity: "))
        df.at[index, "price"] = float(input("Enter New Price: "))
        df.at[index, "threshold"] = int(input("Enter New Threshold: "))

        save_inventory(df, FILE_PATH)
        print("Product updated successfully.")
    else:
        print("Product not found.")

    return df

def delete_product(df):
    """Deletes a product from the inventory."""
    name = input("Enter Product Name to delete: ")

    if name in df["name"].values:
        df = df[df["name"] != name]
        save_inventory(df, FILE_PATH)
        print("Product deleted successfully.")
    else:
        print("Product not found.")

    return df

def generate_low_stock_alerts(df):
    """Displays products that have low stock (less than threshold)."""
    low_stock = df[df["quantity"] < df["threshold"]]
    if low_stock.empty:
        print("No low stock alerts.")
    else:
        print("Low Stock Products:\n", low_stock)

def generate_inventory_report(df):
    """Generates a summary report of the inventory."""
    total_products = df.shape[0]
    total_value = (df["quantity"] * df["price"]).sum()

    print(f"Total Products: {total_products}")
    print(f"Total Inventory Value: ${total_value:.2f}")

def main():
    """Main function to manage inventory operations."""
    inventory = load_inventory(FILE_PATH)

    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. View Products")
        print("3. Update Product")
        print("4. Delete Product")
        print("5. Generate Low Stock Alerts")
        print("6. Generate Inventory Report")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            inventory = add_product(inventory)
        elif choice == "2":
            view_products(inventory)
        elif choice == "3":
            inventory = update_product(inventory)
        elif choice == "4":
            inventory = delete_product(inventory)
        elif choice == "5":
            generate_low_stock_alerts(inventory)
        elif choice == "6":
            generate_inventory_report(inventory)
        elif choice == "7":
            print("Exiting program.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# File path for the inventory CSV file in Google Drive
FILE_PATH = "/content/drive/My Drive/inventory_data.csv"

def load_inventory(file_path):
    """Loads the inventory from a CSV file, or creates a new one if it doesn't exist."""
    if os.path.exists(file_path):
        return pd.read_csv(file_path)
    else:
        df = pd.DataFrame(columns=["name", "category", "quantity", "price", "threshold"])
        df.to_csv(file_path, index=False)
        return df

def save_inventory(df, file_path):
    """Saves the inventory to a CSV file."""
    df.to_csv(file_path, index=False)

def add_product(df):
    """Adds a new product to the inventory or updates the quantity of an existing product."""
    name = input("Enter Product Name: ")

    if name in df["name"].values:
        print("This product already exists.")
        action = input("Do you want to add or delete quantity? (add/delete): ").strip().lower()

        if action == "add":
            quantity_to_add = int(input("Enter quantity to add: "))
            df.loc[df["name"] == name, "quantity"] += quantity_to_add
            print(f"Added {quantity_to_add} to {name}. New quantity: {df.loc[df['name'] == name, 'quantity'].values[0]}")
        elif action == "delete":
            quantity_to_delete = int(input("Enter quantity to delete: "))
            current_quantity = df.loc[df["name"] == name, "quantity"].values[0]
            if current_quantity >= quantity_to_delete:
                df.loc[df["name"] == name, "quantity"] -= quantity_to_delete
                print(f"Deleted {quantity_to_delete} from {name}. New quantity: {df.loc[df['name'] == name, 'quantity'].values[0]}")
            else:
                print("Cannot delete more than the available quantity.")
        else:
            print("Invalid action. Please choose 'add' or 'delete'.")
    else:
        category = input("Enter Product Category: ")
        quantity = int(input("Enter Quantity: "))
        price = float(input("Enter Price: "))
        threshold = int(input("Enter Threshold: "))

        new_product = pd.DataFrame([[name, category, quantity, price, threshold]], columns=df.columns)
        df = pd.concat([df, new_product], ignore_index=True)
        print("Product added successfully.")

    save_inventory(df, FILE_PATH)
    return df

def view_products(df):
    """Displays all products in the inventory."""
    print(df)

def update_product(df):
    """Updates the details of an existing product."""
    name = input("Enter Product Name to update: ")

    if name in df["name"].values:
        index = df[df["name"] == name].index[0]
        print("Existing Details:\n", df.loc[index])

        df.at[index, "name"] = input("Enter New Name: ")
        df.at[index, "category"] = input("Enter New Category: ")
        df.at[index, "quantity"] = int(input("Enter New Quantity: "))
        df.at[index, "price"] = float(input("Enter New Price: "))
        df.at[index, "threshold"] = int(input("Enter New Threshold: "))

        save_inventory(df, FILE_PATH)
        print("Product updated successfully.")
    else:
        print("Product not found.")

    return df

def delete_product(df):
    """Deletes a product from the inventory."""
    name = input("Enter Product Name to delete: ")

    if name in df["name"].values:
        df = df[df["name"] != name]
        save_inventory(df, FILE_PATH)
        print("Product deleted successfully.")
    else:
        print("Product not found.")

    return df

def generate_low_stock_alerts(df):
    """Displays products that have low stock (less than threshold)."""
    low_stock = df[df["quantity"] < df["threshold"]]
    if low_stock.empty:
        print("No low stock alerts.")
    else:
        print("Low Stock Products:\n", low_stock)

def generate_inventory_report(df):
    """Generates a summary report of the inventory."""
    total_products = df.shape[0]
    total_value = (df["quantity"] * df["price"]).sum()

    print(f"Total Products: {total_products}")
    print(f"Total Inventory Value: ${total_value:.2f}")

def main():
    """Main function to manage inventory operations."""
    inventory = load_inventory(FILE_PATH)

    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. View Products")
        print("3. Update Product")
        print("4. Delete Product")
        print("5. Generate Low Stock Alerts")
        print("6. Generate Inventory Report")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            inventory = add_product(inventory)
        elif choice == "2":
            view_products(inventory)
        elif choice == "3":
            inventory = update_product(inventory)
        elif choice == "4":
            inventory = delete_product(inventory)
        elif choice == "5":
            generate_low_stock_alerts(inventory)
        elif choice == "6":
            generate_inventory_report(inventory)
        elif choice == "7":
            print("Exiting program.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# File path for the inventory CSV file in Google Drive
FILE_PATH = "/content/drive/My Drive/inventory_data.csv"

def load_inventory(file_path):
    """Loads the inventory from a CSV file, or creates a new one if it doesn't exist."""
    if os.path.exists(file_path):
        return pd.read_csv(file_path)
    else:
        df = pd.DataFrame(columns=["name", "category", "quantity", "price", "threshold"])
        df.to_csv(file_path, index=False)
        return df

def save_inventory(df, file_path):
    """Saves the inventory to a CSV file."""
    df.to_csv(file_path, index=False)

def add_product(df):
    """Adds a new product to the inventory or updates the quantity of an existing product."""
    name = input("Enter Product Name: ").strip()  # Remove leading/trailing spaces
    normalized_name = name.lower()  # Normalize the name to lowercase for case-insensitive comparison

    # Check if the product already exists (case-insensitive)
    if normalized_name in df["name"].str.lower().values:
        print(f"The product '{name}' already exists.")
        while True:
            action = input("Do you want to add or delete quantity? (add/delete/esc): ").strip().lower()

            if action == "add":
                quantity_to_add = int(input("Enter quantity to add: "))
                df.loc[df["name"].str.lower() == normalized_name, "quantity"] += quantity_to_add
                print(f"Added {quantity_to_add} to {name}. New quantity: {df.loc[df['name'].str.lower() == normalized_name, 'quantity'].values[0]}")
                save_inventory(df, FILE_PATH)
                break
            elif action == "delete":
                quantity_to_delete = int(input("Enter quantity to delete: "))
                current_quantity = df.loc[df["name"].str.lower() == normalized_name, "quantity"].values[0]
                if current_quantity >= quantity_to_delete:
                    df.loc[df["name"].str.lower() == normalized_name, "quantity"] -= quantity_to_delete
                    print(f"Deleted {quantity_to_delete} from {name}. New quantity: {df.loc[df['name'].str.lower() == normalized_name, 'quantity'].values[0]}")
                else:
                    print("Cannot delete more than the available quantity.")
                save_inventory(df, FILE_PATH)
                break
            elif action == "esc":
                print("Returning to the main menu.")
                return df  # Return to the main menu without saving
            else:
                print("Invalid input. Please type 'add', 'delete', or 'esc' to return to the main menu.")
    else:
        category = input("Enter Product Category: ")
        quantity = int(input("Enter Quantity: "))
        price = float(input("Enter Price: "))
        threshold = int(input("Enter Threshold: "))

        new_product = pd.DataFrame([[name, category, quantity, price, threshold]], columns=df.columns)
        df = pd.concat([df, new_product], ignore_index=True)
        print("Product added successfully.")
        save_inventory(df, FILE_PATH)

    return df

def view_products(df):
    """Displays all products in the inventory with their IDs."""
    print(df.to_string())  # Display the DataFrame with indices (IDs)

def update_product(df):
    """Updates specific columns of an existing product based on its ID."""
    view_products(df)  # Show the inventory with IDs
    try:
        product_id = int(input("Enter the ID of the product to update: "))
        if product_id in df.index:
            print("Existing Details:\n", df.loc[product_id])

            # Ask which columns to update
            columns_to_update = input("Enter the columns to update (comma-separated, e.g., 'quantity,price'): ").strip().lower().split(',')
            columns_to_update = [col.strip() for col in columns_to_update]  # Clean up input

            # Validate columns
            valid_columns = ["name", "category", "quantity", "price", "threshold"]
            for col in columns_to_update:
                if col not in valid_columns:
                    print(f"Invalid column: {col}. Valid columns are: {', '.join(valid_columns)}")
                    return df

            # Update the specified columns
            for col in columns_to_update:
                if col == "quantity" or col == "threshold":
                    new_value = int(input(f"Enter new {col}: "))
                elif col == "price":
                    new_value = float(input(f"Enter new {col}: "))
                else:
                    new_value = input(f"Enter new {col}: ")
                df.at[product_id, col] = new_value

            save_inventory(df, FILE_PATH)
            print("Product updated successfully.")
        else:
            print("Invalid ID. Please enter a valid product ID.")
    except ValueError:
        print("Invalid input. Please enter a numeric ID.")

    return df

def delete_product(df):
    """Deletes a product from the inventory based on its ID."""
    view_products(df)  # Show the inventory with IDs
    try:
        product_id = int(input("Enter the ID of the product to delete: "))
        if product_id in df.index:
            product_name = df.loc[product_id, "name"]
            df = df.drop(product_id).reset_index(drop=True)
            save_inventory(df, FILE_PATH)
            print(f"Product '{product_name}' (ID: {product_id}) deleted successfully.")
        else:
            print("Invalid ID. Please enter a valid product ID.")
    except ValueError:
        print("Invalid input. Please enter a numeric ID.")

    return df

def generate_low_stock_alerts(df):
    """Displays products that have low stock (less than threshold)."""
    low_stock = df[df["quantity"] < df["threshold"]]
    if low_stock.empty:
        print("No low stock alerts.")
    else:
        print("Low Stock Products:\n", low_stock.to_string())

def generate_inventory_report(df):
    """Generates a summary report of the inventory."""
    total_products = df.shape[0]
    total_value = (df["quantity"] * df["price"]).sum()

    print(f"Total Products: {total_products}")
    print(f"Total Inventory Value: ${total_value:.2f}")

def main():
    """Main function to manage inventory operations."""
    inventory = load_inventory(FILE_PATH)

    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. View Products")
        print("3. Update Product")
        print("4. Delete Product")
        print("5. Generate Low Stock Alerts")
        print("6. Generate Inventory Report")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            inventory = add_product(inventory)
        elif choice == "2":
            view_products(inventory)
        elif choice == "3":
            inventory = update_product(inventory)
        elif choice == "4":
            inventory = delete_product(inventory)
        elif choice == "5":
            generate_low_stock_alerts(inventory)
        elif choice == "6":
            generate_inventory_report(inventory)
        elif choice == "7":
            print("Exiting program.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import os
import re
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# File paths
INVENTORY_FILE = "/content/drive/My Drive/inventory_data.csv"
WEBSITE_FILE = "/content/drive/My Drive/website_data.csv"

def load_data(file_path, columns):
    """Loads a CSV file or creates a new one if it doesn't exist."""
    if os.path.exists(file_path):
        return pd.read_csv(file_path)
    else:
        df = pd.DataFrame(columns=columns)
        df.to_csv(file_path, index=False)
        return df

def save_data(df, file_path):
    """Saves data to a CSV file."""
    df.to_csv(file_path, index=False)

def view_website_data():
    """Displays website data without low stock warnings"""
    website_df = load_data(WEBSITE_FILE, ["name", "category", "quantity", "price", "threshold"])

    print("\n" + "="*40)
    print("WEBSITE PRODUCTS".center(40))
    print("="*40)
    print(website_df.to_string(index=False))

def restock_website_data():
    """Restocks website data from inventory data when stock falls below threshold."""
    try:
        # Edge Case 9: Missing Files
        if not os.path.exists(WEBSITE_FILE) or not os.path.exists(INVENTORY_FILE):
            raise FileNotFoundError("Inventory or website data file not found.")

        # Load data
        website_df = load_data(WEBSITE_FILE, ["name", "category", "quantity", "price", "threshold"])
        inventory_df = load_data(INVENTORY_FILE, ["name", "category", "quantity", "price", "threshold"])

        # Edge Case 10: Empty Data
        if website_df.empty or inventory_df.empty:
            raise ValueError("Inventory or website data is empty.")

        # Edge Case 3: Missing Columns
        required_columns = ["name", "quantity", "threshold"]
        for col in required_columns:
            if col not in website_df.columns or col not in inventory_df.columns:
                raise ValueError(f"Missing required column: {col}")

        # Edge Case 4: Duplicate Products
        if website_df["name"].duplicated().any() or inventory_df["name"].duplicated().any():
            raise ValueError("Duplicate product names found in inventory or website data.")

        restocked = False
        for index, row in website_df.iterrows():
            try:
                # Edge Case 1: Negative Quantity or Threshold
                if row["quantity"] < 0 or row["threshold"] < 0:
                    raise ValueError(f"Negative quantity or threshold for product: {row['name']}")

                # Edge Case 2: Zero Threshold
                if row["threshold"] == 0:
                    raise ValueError(f"Threshold is zero for product: {row['name']}")

                # Skip if quantity is already above threshold
                if row["quantity"] >= row["threshold"]:
                    continue

                # Edge Case 5: Case Sensitivity in Product Names
                product_name = row["name"].lower().strip()
                inventory_match = inventory_df[inventory_df["name"].str.lower().str.strip() == product_name]

                if inventory_match.empty:
                    raise ValueError(f"Product {row['name']} not found in inventory.")

                available_stock = inventory_match.iloc[0]["quantity"]
                restock_amount = row["threshold"] - row["quantity"]

                # Edge Case 7: Data Type Mismatch
                if not isinstance(available_stock, (int, float)) or not isinstance(restock_amount, (int, float)):
                    raise ValueError(f"Invalid data type for quantity or threshold in product: {row['name']}")

                # Check if inventory quantity will go below zero after restocking
                new_inventory_quantity = available_stock - restock_amount
                if new_inventory_quantity < 0:
                    raise ValueError(f"Restocking {row['name']} will cause inventory quantity to go below zero.")

                # Perform restocking
                website_df.at[index, "quantity"] += restock_amount
                inventory_df.loc[inventory_df["name"].str.lower().str.strip() == product_name, "quantity"] -= restock_amount
                print(f"Restocked {restock_amount} units of {row['name']} from inventory.")
                restocked = True

            except Exception as e:
                print(f"Error: {str(e)}")
                raise  # Stop execution on error

        if restocked:
            save_data(website_df, WEBSITE_FILE)
            save_data(inventory_df, INVENTORY_FILE)
    except Exception as e:
        print(f"Fatal Error: {str(e)}")
        raise  # Stop execution on fatal error

def is_valid_name_or_category(value):
    """Validates that the name or category starts with a letter."""
    return bool(re.match("^[A-Za-z]", value))

def add_product(df):
    """Adds a new product to the inventory or updates quantity of existing product."""
    name = input("Enter Product Name: ").strip()

    while not is_valid_name_or_category(name):
        print("Invalid product name. Name must start with a letter.")
        name = input("Enter Product Name: ").strip()

    normalized_name = name.lower()

    if normalized_name in df["name"].str.lower().values:
        print(f"The product '{name}' already exists.")
        while True:
            action = input("Do you want to add or delete quantity? (add/delete/esc): ").strip().lower()

            if action == "add":
                quantity_to_add = int(input("Enter quantity to add: "))
                df.loc[df["name"].str.lower() == normalized_name, "quantity"] += quantity_to_add
                print(f"Added {quantity_to_add} to {name}. New quantity: {df.loc[df['name'].str.lower() == normalized_name, 'quantity'].values[0]}")
                save_data(df, INVENTORY_FILE)
                break
            elif action == "delete":
                quantity_to_delete = int(input("Enter quantity to delete: "))
                current_quantity = df.loc[df["name"].str.lower() == normalized_name, "quantity"].values[0]
                if current_quantity >= quantity_to_delete:
                    df.loc[df["name"].str.lower() == normalized_name, "quantity"] -= quantity_to_delete
                    print(f"Deleted {quantity_to_delete} from {name}. New quantity: {current_quantity - quantity_to_delete}")
                else:
                    print("Cannot delete more than available quantity.")
                save_data(df, INVENTORY_FILE)
                break
            elif action == "esc":
                return df
            else:
                print("Invalid input. Please type 'add', 'delete', or 'esc'.")
    else:
        category = input("Enter Product Category: ")
        while not is_valid_name_or_category(category):
            print("Invalid category. Must start with a letter.")
            category = input("Enter Product Category: ").strip()

        quantity = -1
        while quantity < 0:
            try:
                quantity = int(input("Enter Quantity: "))
                if quantity < 0: raise ValueError
            except ValueError:
                print("Quantity must be a positive integer.")

        price = -1.0
        while price < 0:
            try:
                price = float(input("Enter Price: "))
                if price < 0: raise ValueError
            except ValueError:
                print("Price must be a positive number.")

        threshold = -1
        while threshold < 0:
            try:
                threshold = int(input("Enter Threshold: "))
                if threshold < 0: raise ValueError
            except ValueError:
                print("Threshold must be a positive integer.")

        new_product = pd.DataFrame([[name, category, quantity, price, threshold]],
                                 columns=df.columns)
        df = pd.concat([df, new_product], ignore_index=True)
        print("Product added successfully.")
        save_data(df, INVENTORY_FILE)

    return df

def view_products(df):
    """Displays all products in the inventory with their IDs."""
    print("\n" + "="*40)
    print("INVENTORY PRODUCTS".center(40))
    print("="*40)
    # Reset index to show IDs explicitly
    print(df.reset_index(drop=False).to_string(index=False, header=True))

def update_product(df):
    """Updates specific columns of an existing product."""
    view_products(df)  # Show the inventory with IDs
    try:
        product_id = int(input("Enter the ID of the product to update: "))
        if product_id in df.index:
            while True:
                print("\nCurrent Details:\n", df.loc[product_id])
                column = input("Enter column to update (name/category/quantity/price/threshold): ").lower()

                valid_columns = ["name", "category", "quantity", "price", "threshold"]
                if column not in valid_columns:
                    print(f"Invalid column. Valid options: {', '.join(valid_columns)}")
                    continue

                new_value = None
                if column in ["quantity", "threshold"]:
                    while True:
                        try:
                            new_value = int(input(f"Enter new {column}: "))
                            if new_value >= 0: break
                            print("Must be non-negative integer")
                        except ValueError:
                            print("Invalid integer input")
                elif column == "price":
                    while True:
                        try:
                            new_value = float(input(f"Enter new {column}: "))
                            if new_value >= 0: break
                            print("Must be non-negative number")
                        except ValueError:
                            print("Invalid number input")
                else:
                    while True:
                        new_value = input(f"Enter new {column}: ").strip()
                        if is_valid_name_or_category(new_value):
                            break
                        print(f"Invalid {column}. Must start with letter")

                df.at[product_id, column] = new_value
                save_data(df, INVENTORY_FILE)
                print(f"{column.capitalize()} updated successfully.")

                if input("Update another field? (y/n): ").lower() != 'y':
                    break
        else:
            print("Invalid ID")
    except ValueError:
        print("Invalid input. Please enter numeric ID.")
    return df

def delete_product(df):
    """Deletes a product from the inventory."""
    view_products(df)
    try:
        product_id = int(input("Enter the ID of the product to delete: "))
        if product_id in df.index:
            confirm = input(f"Delete product '{df.loc[product_id, 'name']}'? (y/n): ")
            if confirm.lower() == 'y':
                df = df.drop(product_id).reset_index(drop=True)
                save_data(df, INVENTORY_FILE)
                print("Product deleted successfully.")
        else:
            print("Invalid ID")
    except ValueError:
        print("Invalid input. Please enter numeric ID.")
    return df

def generate_low_stock_alerts():
    """Displays low stock items for either inventory or website."""
    print("\nCheck low stock for:")
    print("1. Inventory")
    print("2. Website")
    choice = input("Enter your choice (1-2): ").strip()

    if choice == '1':
        df = load_data(INVENTORY_FILE, ["name", "category", "quantity", "price", "threshold"])
        low_stock = df[df['quantity'] < df['threshold']]
        if low_stock.empty:
            print("\nNo low stock items in inventory.")
        else:
            print("\n" + "="*40)
            print("INVENTORY LOW STOCK ALERTS".center(40))
            print("="*40)
            print(low_stock[['name', 'quantity', 'threshold']].to_string(index=False))
    elif choice == '2':
        df = load_data(WEBSITE_FILE, ["name", "category", "quantity", "price", "threshold"])
        low_stock = df[df['quantity'] < df['threshold']]
        if low_stock.empty:
            print("\nNo low stock items on website.")
        else:
            print("\n" + "="*40)
            print("WEBSITE LOW STOCK ALERTS".center(40))
            print("="*40)
            print(low_stock[['name', 'quantity', 'threshold']].to_string(index=False))
    else:
        print("Invalid choice.")

def generate_inventory_report(df):
    """Generates summary report of inventory."""
    total_value = (df['quantity'] * df['price']).sum()
    print("\nINVENTORY REPORT:")
    print(f"Total Products: {len(df)}")
    print(f"Total Inventory Value: ${total_value:,.2f}")

def main():
    inventory_df = load_data(INVENTORY_FILE, ["name", "category", "quantity", "price", "threshold"])

    while True:
        print("\n" + "="*40)
        print("INVENTORY MANAGEMENT SYSTEM".center(40))
        print("="*40)
        print("1. Add Product")
        print("2. View Inventory Products")
        print("3. Update Product")
        print("4. Delete Product")
        print("5. Check Low Stock Alerts")
        print("6. Inventory Report")
        print("7. Restock Website Products")
        print("8. View Website Products")
        print("9. Exit")

        choice = input("\nEnter your choice (1-9): ").strip()

        if choice == '1':
            inventory_df = add_product(inventory_df)
        elif choice == '2':
            view_products(inventory_df)
        elif choice == '3':
            inventory_df = update_product(inventory_df)
        elif choice == '4':
            inventory_df = delete_product(inventory_df)
        elif choice == '5':
            generate_low_stock_alerts()
        elif choice == '6':
            generate_inventory_report(inventory_df)
        elif choice == '7':
            restock_website_data()
        elif choice == '8':
            view_website_data()
        elif choice == '9':
            print("\nExiting system. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1-9.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import os
import re
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# File paths
INVENTORY_FILE = "/content/drive/My Drive/inventory_data.csv"
WEBSITE_FILE = "/content/drive/My Drive/website_data.csv"

def load_data(file_path, columns):
    """Loads a CSV file or creates a new one if it doesn't exist."""
    if os.path.exists(file_path):
        return pd.read_csv(file_path)
    else:
        df = pd.DataFrame(columns=columns)
        df.to_csv(file_path, index=False)
        return df

def save_data(df, file_path):
    """Saves data to a CSV file."""
    df.to_csv(file_path, index=False)

def is_valid_name_or_category(value):
    """Validates that the name or category starts with a letter."""
    return bool(re.match("^[A-Za-z]", value))

def view_products(df, title):
    """Displays all products in the given DataFrame with their IDs."""
    print("\n" + "="*40)
    print(title.center(40))
    print("="*40)
    # Reset index to show IDs explicitly
    print(df.reset_index(drop=False).to_string(index=False, header=True))

def add_product(df, file_path):
    """Adds a new product to the given DataFrame or updates quantity of existing product."""
    name = input("Enter Product Name: ").strip()

    while not is_valid_name_or_category(name):
        print("Invalid product name. Name must start with a letter.")
        name = input("Enter Product Name: ").strip()

    normalized_name = name.lower()

    if normalized_name in df["name"].str.lower().values:
        print(f"The product '{name}' already exists.")
        while True:
            action = input("Do you want to add or delete quantity? (add/delete/esc): ").strip().lower()

            if action == "add":
                quantity_to_add = int(input("Enter quantity to add: "))
                df.loc[df["name"].str.lower() == normalized_name, "quantity"] += quantity_to_add
                print(f"Added {quantity_to_add} to {name}. New quantity: {df.loc[df['name'].str.lower() == normalized_name, 'quantity'].values[0]}")
                save_data(df, file_path)
                break
            elif action == "delete":
                quantity_to_delete = int(input("Enter quantity to delete: "))
                current_quantity = df.loc[df["name"].str.lower() == normalized_name, "quantity"].values[0]
                if current_quantity >= quantity_to_delete:
                    df.loc[df["name"].str.lower() == normalized_name, "quantity"] -= quantity_to_delete
                    print(f"Deleted {quantity_to_delete} from {name}. New quantity: {current_quantity - quantity_to_delete}")
                else:
                    print("Cannot delete more than available quantity.")
                save_data(df, file_path)
                break
            elif action == "esc":
                return df
            else:
                print("Invalid input. Please type 'add', 'delete', or 'esc'.")
    else:
        category = input("Enter Product Category: ")
        while not is_valid_name_or_category(category):
            print("Invalid category. Must start with a letter.")
            category = input("Enter Product Category: ").strip()

        quantity = -1
        while quantity < 0:
            try:
                quantity = int(input("Enter Quantity: "))
                if quantity < 0: raise ValueError
            except ValueError:
                print("Quantity must be a positive integer.")

        price = -1.0
        while price < 0:
            try:
                price = float(input("Enter Price: "))
                if price < 0: raise ValueError
            except ValueError:
                print("Price must be a positive number.")

        threshold = -1
        while threshold < 0:
            try:
                threshold = int(input("Enter Threshold: "))
                if threshold < 0: raise ValueError
            except ValueError:
                print("Threshold must be a positive integer.")

        new_product = pd.DataFrame([[name, category, quantity, price, threshold]],
                                 columns=df.columns)
        df = pd.concat([df, new_product], ignore_index=True)
        print("Product added successfully.")
        save_data(df, file_path)

    return df

def update_product(df, file_path):
    """Updates specific columns of an existing product in the given DataFrame."""
    view_products(df, "CURRENT PRODUCTS")
    try:
        product_id = int(input("Enter the ID of the product to update: "))
        if product_id in df.index:
            while True:
                print("\nCurrent Details:\n", df.loc[product_id])
                column = input("Enter column to update (name/category/quantity/price/threshold): ").lower()

                valid_columns = ["name", "category", "quantity", "price", "threshold"]
                if column not in valid_columns:
                    print(f"Invalid column. Valid options: {', '.join(valid_columns)}")
                    continue

                new_value = None
                if column in ["quantity", "threshold"]:
                    while True:
                        try:
                            new_value = int(input(f"Enter new {column}: "))
                            if new_value >= 0: break
                            print("Must be non-negative integer")
                        except ValueError:
                            print("Invalid integer input")
                elif column == "price":
                    while True:
                        try:
                            new_value = float(input(f"Enter new {column}: "))
                            if new_value >= 0: break
                            print("Must be non-negative number")
                        except ValueError:
                            print("Invalid number input")
                else:
                    while True:
                        new_value = input(f"Enter new {column}: ").strip()
                        if is_valid_name_or_category(new_value):
                            break
                        print(f"Invalid {column}. Must start with letter")

                df.at[product_id, column] = new_value
                save_data(df, file_path)
                print(f"{column.capitalize()} updated successfully.")

                if input("Update another field? (y/n): ").lower() != 'y':
                    break
        else:
            print("Invalid ID")
    except ValueError:
        print("Invalid input. Please enter numeric ID.")
    return df

def delete_product(df, file_path):
    """Deletes a product from the given DataFrame."""
    view_products(df, "CURRENT PRODUCTS")
    try:
        product_id = int(input("Enter the ID of the product to delete: "))
        if product_id in df.index:
            confirm = input(f"Delete product '{df.loc[product_id, 'name']}'? (y/n): ")
            if confirm.lower() == 'y':
                df = df.drop(product_id).reset_index(drop=True)
                save_data(df, file_path)
                print("Product deleted successfully.")
        else:
            print("Invalid ID")
    except ValueError:
        print("Invalid input. Please enter numeric ID.")
    return df

def generate_low_stock_alerts(df, title):
    """Displays low stock items for the given DataFrame."""
    low_stock = df[df['quantity'] < df['threshold']]
    if low_stock.empty:
        print(f"\nNo low stock items in {title}.")
    else:
        print("\n" + "="*40)
        print(f"{title.upper()} LOW STOCK ALERTS".center(40))
        print("="*40)
        print(low_stock[['name', 'quantity', 'threshold']].to_string(index=False))

def generate_inventory_report(df):
    """Generates summary report of inventory."""
    total_value = (df['quantity'] * df['price']).sum()
    print("\nINVENTORY REPORT:")
    print(f"Total Products: {len(df)}")
    print(f"Total Inventory Value: ${total_value:,.2f}")

def restock_website_data(inventory_df, website_df):
    """Restocks website data from inventory data when stock falls below threshold."""
    restocked = False
    for index, row in website_df.iterrows():
        if row["quantity"] < row["threshold"]:
            product_name = row["name"].lower().strip()
            inventory_match = inventory_df[inventory_df["name"].str.lower().str.strip() == product_name]

            if not inventory_match.empty:
                available_stock = inventory_match.iloc[0]["quantity"]
                restock_amount = row["threshold"] - row["quantity"]

                # Perform restocking without checking if inventory quantity goes below zero
                website_df.at[index, "quantity"] += restock_amount
                inventory_df.loc[inventory_df["name"].str.lower().str.strip() == product_name, "quantity"] -= restock_amount
                print(f"Restocked {restock_amount} units of {row['name']} from inventory.")
                restocked = True
            else:
                print(f"Product {row['name']} not found in inventory.")

    if restocked:
        save_data(website_df, WEBSITE_FILE)
        save_data(inventory_df, INVENTORY_FILE)
        print("Restocking completed. Inventory and website data updated.")
    else:
        print("No restocking needed.")

    return inventory_df, website_df

def main():
    inventory_df = load_data(INVENTORY_FILE, ["name", "category", "quantity", "price", "threshold"])
    website_df = load_data(WEBSITE_FILE, ["name", "category", "quantity", "price", "threshold"])

    while True:
        print("\n" + "="*40)
        print("INVENTORY MANAGEMENT SYSTEM".center(40))
        print("="*40)
        print("1. Add Product to Inventory")
        print("2. View Inventory Products")
        print("3. Update Inventory Product")
        print("4. Delete Inventory Product")
        print("5. Check Low Stock Alerts (Inventory)")
        print("6. Inventory Report")
        print("7. Restock Website Products")
        print("8. View Website Products")
        print("9. Add Product to Website")
        print("10. Update Website Product")
        print("11. Delete Website Product")
        print("12. Check Low Stock Alerts (Website)")
        print("13. Exit")

        choice = input("\nEnter your choice (1-13): ").strip()

        if choice == '1':
            inventory_df = add_product(inventory_df, INVENTORY_FILE)
        elif choice == '2':
            view_products(inventory_df, "INVENTORY PRODUCTS")
        elif choice == '3':
            inventory_df = update_product(inventory_df, INVENTORY_FILE)
        elif choice == '4':
            inventory_df = delete_product(inventory_df, INVENTORY_FILE)
        elif choice == '5':
            generate_low_stock_alerts(inventory_df, "inventory")
        elif choice == '6':
            generate_inventory_report(inventory_df)
        elif choice == '7':
            inventory_df, website_df = restock_website_data(inventory_df, website_df)
        elif choice == '8':
            view_products(website_df, "WEBSITE PRODUCTS")
        elif choice == '9':
            website_df = add_product(website_df, WEBSITE_FILE)
        elif choice == '10':
            website_df = update_product(website_df, WEBSITE_FILE)
        elif choice == '11':
            website_df = delete_product(website_df, WEBSITE_FILE)
        elif choice == '12':
            generate_low_stock_alerts(website_df, "website")
        elif choice == '13':
            print("\nExiting system. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1-13.")

if __name__ == "__main__":
    main()
```

```python
import re
import json
import random
from datetime import datetime, timedelta
from tabulate import tabulate

class FlightBookingSystem:
    def __init__(self):
        self.flights = self.load_flights()
        self.bookings = self.load_bookings()
        self.waitlist = []

        if not self.flights:
            self._initialize_us_flights()
            self.save_flights()

    def _initialize_us_flights(self):
        """Create 50 initial US domestic flights with enhanced validation."""
        us_cities = [
            'New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix',
            'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose',
            'Austin', 'Jacksonville', 'Fort Worth', 'Columbus', 'San Francisco',
            'Charlotte', 'Indianapolis', 'Seattle', 'Denver', 'Washington DC',
            'Boston', 'El Paso', 'Detroit', 'Nashville', 'Memphis', 'Portland',
            'Oklahoma City', 'Las Vegas', 'Baltimore', 'Louisville', 'Milwaukee',
            'Albuquerque', 'Tucson', 'Fresno', 'Sacramento', 'Kansas City',
            'Atlanta', 'Miami', 'Raleigh', 'Omaha', 'Long Beach', 'Virginia Beach',
            'Colorado Springs', 'Oakland', 'Minneapolis', 'Tulsa', 'Tampa',
            'Arlington', 'New Orleans', 'Wichita', 'Cleveland', 'Bakersfield',
            'Aurora', 'Anaheim', 'Honolulu', 'Santa Ana', 'Riverside',
            'Corpus Christi', 'Lexington', 'Henderson', 'Stockton', 'Saint Paul'
        ]

        existing_numbers = set()
        for i in range(50):  # Maintain about 50 flights
            flight_num = f"AA{i+1:03d}"

            # Ensure unique flight number before creating flight
            while flight_num in existing_numbers:
                flight_num = f"AA{random.randint(100, 999)}"

            existing_numbers.add(flight_num)

            origin = random.choice(us_cities)
            destination = random.choice([c for c in us_cities if c != origin])
            capacity = random.randint(150, 500)
            flight = {
                'flight_number': flight_num,
                'origin': origin,
                'destination': destination,
                'departure_datetime': (datetime.now() + timedelta(days=random.randint(1,180))).strftime('%Y-%m-%d %I:%M %p'),
                'landing_datetime': (datetime.now() + timedelta(days=random.randint(1,180), hours=random.randint(1,6))).strftime('%Y-%m-%d %I:%M %p'),
                'capacity': capacity,
                'passengers': random.randint(0, capacity),
                'status': 'Scheduled'
            }

            self.flights.append(flight)

    def _validate_flight_details(self, flight_number, origin, destination,
                               departure_datetime, landing_datetime, capacity):
        """Validate flight details for edge cases."""
        if not re.match(r'^[A-Za-z]{2}\d{3}$', flight_number):
            return False
        if not re.match(r'^[A-Za-z\s\-]+$', origin) or not re.match(r'^[A-Za-z\s\-]+$', destination):
            return False
        if origin.title() == destination.title():
            return False
        now = datetime.now()
        if departure_datetime < now or landing_datetime < now:
            return False
        max_date = now + timedelta(days=180)
        if departure_datetime > max_date or landing_datetime > max_date:
            return False
        if landing_datetime.date() < departure_datetime.date():
            return False
        if (landing_datetime.date() == departure_datetime.date() and
            landing_datetime.time() <= departure_datetime.time()):
            return False
        if (landing_datetime - departure_datetime) > timedelta(hours=24):
            return False
        if not isinstance(capacity, int) or capacity <= 0 or capacity > 500:
            return False
        return True

    def add_flight(self, flight_number, origin, destination, departure_datetime, landing_datetime, capacity, passengers):
        """Add a new flight to the system."""
        flight_number = flight_number.upper()

        # Check for duplicate flight number
        if any(f['flight_number'] == flight_number for f in self.flights):
            return "Flight number already exists"

        # Validate flight details
        origin = origin.title()
        destination = destination.title()

        if not self._validate_flight_details(flight_number, origin, destination,
                                           departure_datetime, landing_datetime, capacity):
            return "Invalid flight details"

        if passengers < 0 or passengers > capacity:
            return f"Passengers must be between 0-{capacity}"

        flight = {
            'flight_number': flight_number,
            'origin': origin,
            'destination': destination,
            'departure_datetime': departure_datetime.strftime('%Y-%m-%d %I:%M %p'),
            'landing_datetime': landing_datetime.strftime('%Y-%m-%d %I:%M %p'),
            'capacity': capacity,
            'passengers': passengers,
            'status': 'Scheduled'
        }
        self.flights.append(flight)
        self.save_flights()
        return f"Flight {flight_number} added successfully"

    def delete_flight(self, flight_number):
        """Delete a flight from the system."""
        if not self.flights:
            return "The table is already empty."

        flight_number = flight_number.upper()
        for flight in self.flights:
            if flight['flight_number'] == flight_number:
                self.bookings = [b for b in self.bookings if b['flight_number'] != flight_number]
                self.waitlist = [w for w in self.waitlist if w['flight_number'] != flight_number]
                self.flights.remove(flight)
                self.save_flights()
                self.save_bookings()
                return f"Flight {flight_number} deleted successfully"
        return f"Flight {flight_number} not found"

    def display_flights(self):
        """Display all flights in a tabular format."""
        if not self.flights:
            print("No flights available.")
            return

        table = []
        for flight in self.flights:
            available = flight['capacity'] - flight['passengers']
            table.append([
                flight['flight_number'],
                flight['origin'],
                flight['destination'],
                flight['departure_datetime'],
                flight['landing_datetime'],
                flight['capacity'],
                flight['passengers'],
                f"{available} seats",
                "üü¢" if available > 0 else "üî¥"
            ])

        print(tabulate(table, headers=[
            "Flight", "From", "To", "Departure", "Arrival",
            "Capacity", "Passengers", "Available", "Status"
        ], tablefmt="grid"))

    def book_flight(self, flight_number, passenger_name, email, phone):
        """Book a flight for a passenger."""
        passenger_name = passenger_name.strip().title()
        if not passenger_name:
            return "Passenger name cannot be empty"
        if not re.match(r'^[\w\.-]+@gmail\.com$', email.strip().lower()):
            return "Invalid email - must be @gmail.com"
        if not re.match(r'^\d{10}$', phone.strip()):
            return "Phone must be 10 digits"

        flight_number = flight_number.upper()
        for flight in self.flights:
            if flight['flight_number'] == flight_number:
                if flight['passengers'] < flight['capacity']:
                    booking_id = len(self.bookings) + 1
                    self.bookings.append({
                        'booking_id': booking_id,
                        'flight_number': flight_number,
                        'passenger': passenger_name,
                        'email': email.lower(),
                        'phone': phone.strip(),
                        'status': 'Confirmed'
                    })
                    flight['passengers'] += 1
                    self.save_flights()
                    self.save_bookings()
                    return f"Booking confirmed! ID: {booking_id}"
                else:
                    self.waitlist.append({
                        'flight_number': flight_number,
                        'passenger': passenger_name,
                        'email': email,
                        'phone': phone
                    })
                    return "Flight full - added to waitlist"
        return "Flight not found"

    def cancel_booking(self, booking_id):
        """Cancel a booking and process the waitlist."""
        try:
            booking_id = int(booking_id)
        except ValueError:
            return "Invalid booking ID"

        for booking in self.bookings:
            if booking['booking_id'] == booking_id:
                for flight in self.flights:
                    if flight['flight_number'] == booking['flight_number']:
                        flight['passengers'] = max(0, flight['passengers'] - 1)
                        self.save_flights()
                        break
                self._process_waitlist(booking['flight_number'])
                self.bookings.remove(booking)
                self.save_bookings()
                return f"Cancelled booking {booking_id}"
        return "Booking not found"

    def _process_waitlist(self, flight_number):
        """Process the waitlist for a flight."""
        available = next((f['capacity'] - f['passengers']
                        for f in self.flights
                        if f['flight_number'] == flight_number), 0)

        while available > 0 and self.waitlist:
            passenger = next((p for p in self.waitlist
                            if p['flight_number'] == flight_number), None)
            if passenger:
                self.book_flight(flight_number,
                               passenger['passenger'],
                               passenger['email'],
                               passenger['phone'])
                self.waitlist.remove(passenger)
                available -= 1

    def find_trips(self, origin, destination, date):
        """Find trips based on origin, destination, and date."""
        origin = origin.strip().title()
        destination = destination.strip().title()
        date_str = date.strftime('%Y-%m-%d')
        matching = []

        for flight in self.flights:
            if (flight['origin'].title() == origin and
                flight['destination'].title() == destination and
                flight['departure_datetime'].startswith(date_str) and
                flight['status'] == 'Scheduled'):
                matching.append(flight)
        return matching

    def generate_performance_report(self):
        """Generate a performance report for all flights."""
        report = []
        for flight in self.flights:
            occupancy_rate = (flight['passengers'] / flight['capacity']) * 100
            report.append([
                flight['flight_number'],
                flight['origin'],
                flight['destination'],
                flight['departure_datetime'],
                flight['capacity'],
                flight['passengers'],
                f"{occupancy_rate:.2f}%"
            ])

        print(tabulate(report, headers=[
            "Flight", "From", "To", "Departure", "Capacity", "Passengers", "Occupancy Rate"
        ], tablefmt="grid"))

    def update_flight_status(self, flight_number, status):
        """Update the status of a flight (e.g., Delayed, Cancelled)."""
        flight_number = flight_number.upper()
        for flight in self.flights:
            if flight['flight_number'] == flight_number:
                flight['status'] = status
                self.save_flights()
                return f"Flight {flight_number} status updated to {status}"
        return f"Flight {flight_number} not found"

    def save_flights(self):
        """Save flights to a JSON file."""
        with open('flights.json', 'w') as f:
            json.dump(self.flights, f, indent=4)

    def load_flights(self):
        """Load flights from a JSON file."""
        try:
            with open('flights.json', 'r') as f:
                return json.load(f)
        except (FileNotFoundError, json.JSONDecodeError):
            return []

    def save_bookings(self):
        """Save bookings to a JSON file."""
        with open('bookings.json', 'w') as f:
            json.dump(self.bookings, f, indent=4)

    def load_bookings(self):
        """Load bookings from a JSON file."""
        try:
            with open('bookings.json', 'r') as f:
                return json.load(f)
        except (FileNotFoundError, json.JSONDecodeError):
            return []

def validate_date(date_str):
    """Validate a date string in YYYY-MM-DD format."""
    try:
        datetime.strptime(date_str, "%Y-%m-%d")
        return True
    except ValueError:
        return False

def main():
    system = FlightBookingSystem()

    while True:
        print("\nFlight Booking System")
        print("1. Add Flight\n2. View Flights\n3. Find Trips\n4. Delete Flight")
        print("5. Book Flight\n6. Cancel Booking\n7. Generate Performance Report")
        print("8. Update Flight Status\n9. Exit")
        choice = input("Enter choice: ").strip()

        if choice == "1":
            while True:
                flight_number = input("Flight Number (AA123): ").strip().upper()
                if not re.match(r'^[A-Z]{2}\d{3}$', flight_number):
                    print("Invalid format! Must be 2 letters + 3 digits (e.g., AA123)")
                    continue
                if any(f['flight_number'] == flight_number for f in system.flights):
                    print("Flight number already exists!")
                    continue
                break

            origin = input("Origin: ").strip().title()
            destination = input("Destination: ").strip().title()

            try:
                departure = datetime.strptime(input("Departure (YYYY-MM-DD HH:MM AM/PM): "), "%Y-%m-%d %I:%M %p")
                landing = datetime.strptime(input("Landing (YYYY-MM-DD HH:MM AM/PM): "), "%Y-%m-%d %I:%M %p")
                capacity = int(input("Capacity: "))
                passengers = int(input("Current Passengers: "))
            except ValueError:
                print("Invalid input format")
                continue

            result = system.add_flight(flight_number, origin, destination,
                                      departure, landing, capacity, passengers)
            print(result)

        elif choice == "2":
            system.display_flights()

        elif choice == "3":
            origin = input("From: ").strip().title()
            destination = input("To: ").strip().title()
            date_str = input("Date (YYYY-MM-DD): ").strip()

            if validate_date(date_str):
                date = datetime.strptime(date_str, "%Y-%m-%d")
                flights = system.find_trips(origin, destination, date)
                if flights:
                    print("\nAvailable Flights:")
                    for idx, f in enumerate(flights, 1):
                        print(f"{idx}. {f['flight_number']} | {f['departure_datetime']} | Seats: {f['capacity'] - f['passengers']}")
                else:
                    print("No flights found")
            else:
                print("Invalid date format")

        elif choice == "4":
            flight_number = input("Enter flight number to delete: ").strip()
            print(system.delete_flight(flight_number))

        elif choice == "5":
            origin = input("From: ").strip().title()
            destination = input("To: ").strip().title()
            date_str = input("Date (YYYY-MM-DD): ").strip()

            if validate_date(date_str):
                date = datetime.strptime(date_str, "%Y-%m-%d")
                flights = system.find_trips(origin, destination, date)
                if flights:
                    print("\nAvailable Flights:")
                    for idx, f in enumerate(flights, 1):
                        print(f"{idx}. {f['flight_number']} | {f['departure_datetime']} | Seats: {f['capacity'] - f['passengers']}")

                    try:
                        flight_choice = int(input("Select flight: ")) - 1
                        selected = flights[flight_choice]
                        num_passengers = int(input("Number of passengers: "))

                        for p in range(num_passengers):
                            print(f"\nPassenger {p+1}:")
                            name = input("Name: ").strip().title()
                            email = input("Email: ").strip()
                            phone = input("Phone: ").strip()
                            result = system.book_flight(selected['flight_number'], name, email, phone)
                            print(result)
                    except (ValueError, IndexError):
                        print("Invalid selection")
                else:
                    print("No flights found")
            else:
                print("Invalid date")

        elif choice == "6":
            booking_id = input("Enter booking ID to cancel: ")
            print(system.cancel_booking(booking_id))

        elif choice == "7":
            system.generate_performance_report()

        elif choice == "8":
            flight_number = input("Enter flight number: ").strip().upper()
            status = input("Enter new status (e.g., Delayed, Cancelled): ").strip().title()
            print(system.update_flight_status(flight_number, status))

        elif choice == "9":
            print("Exiting system.")
            break

        else:
            print("Invalid choice")

if __name__ == "__main__":
    main()
```
