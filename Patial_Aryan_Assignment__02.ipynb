```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd

# Set Chrome options
chrome_options = Options()
chrome_options.add_argument('--headless')  # Run in headless mode
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36")
chrome_options.add_argument('--blink-settings=imagesEnabled=false')  # Disable images

# Initialize WebDriver
service = Service(ChromeDriverManager().install())
wd = webdriver.Chrome(service=service, options=chrome_options)

# Navigate to IMDb reviews page
wd.get("https://www.imdb.com/title/tt15398776/reviews/")

# Scroll to the bottom of the page to load the "All" button
wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(2)  # Short wait for the button to load

# Click the "All" button (if it exists)
try:
    all_button = WebDriverWait(wd, 5).until(
        EC.element_to_be_clickable((By.XPATH, "//button//span[text()='All']"))
    )
    wd.execute_script("arguments[0].click();", all_button)
    time.sleep(5)  # Short wait for reviews to load
except Exception as e:
    print(f"Error clicking 'All' button: {e}")

# Extract reviews in batches
reviews_data = []
max_reviews = 1000  # Set the desired number of reviews
batch_size = 25  # Number of reviews loaded per click
wait_time = 5  # Time to wait between clicks in seconds
retry_limit = 3  # Number of retries if "25 more" fails

while len(reviews_data) < max_reviews:
    # Extract reviews
    reviews = WebDriverWait(wd, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//div[@class='ipc-html-content-inner-div']"))
    )
    for review in reviews:
        if len(reviews_data) >= max_reviews:
            break
        reviews_data.append(review.text.strip())
        print(f"Collected review {len(reviews_data)}: {review.text.strip()[:50]}...")  # Print a snippet of the review

    # Check if we need to load more reviews
    if len(reviews_data) < max_reviews:
        retries = 0
        while retries < retry_limit:
            try:
                # Scroll to the bottom of the page to load the "25 more" button
                wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(2)  # Short wait for the button to load

                # Click the "25 more" button
                load_more_button = WebDriverWait(wd, 5).until(
                    EC.element_to_be_clickable((By.XPATH, "//button//span[contains(text(), '25 more')]"))
                )
                wd.execute_script("arguments[0].click();", load_more_button)
                time.sleep(wait_time)  # Wait for the new reviews to load
                break  # Exit the retry loop if successful
            except Exception as e:
                print(f"Attempt {retries + 1}: Error clicking '25 more' button: {e}")
                retries += 1
                time.sleep(5)  # Wait before retrying
        else:
            print("Failed to load more reviews after multiple attempts. Exiting.")
            break

# Close the browser session
wd.quit()

# Convert to DataFrame and save to CSV
df = pd.DataFrame(reviews_data, columns=["review"])
df.to_csv('imdb_reviews.csv', index=False)

print(f"Reviews scraped: {len(reviews_data)}. Saved to 'imdb_reviews.csv'.")
```

Write a python program to **conduct syntax and structure analysis of the clean text** you just saved above. The syntax and structure analysis includes:

(1) **Parts of Speech (POS) Tagging:** Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) **Constituency Parsing and Dependency Parsing:** print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) **Named Entity Recognition:** Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
# Your code here

#1 Parts of Speech(POS)
import spacy

spacy_eng_model = spacy.load("en_core_web_sm")

total_counts = {
    "nouns": 0,
    "verbs": 0,
    "adjectives": 0,
    "adverbs": 0
}

def pos(feature):
    pos_model = spacy_eng_model(feature)  # main POS happening here
    counts = {
        "nouns": 0,
        "verbs": 0,
        "adjectives": 0,
        "adverbs": 0
    }

    for words in pos_model:
        if words.pos_ == "NOUN":
            counts["nouns"] += 1
        elif words.pos_ == "VERB":
            counts["verbs"] += 1
        elif words.pos_ == "ADJ":
            counts["adjectives"] += 1
        elif words.pos_ == "ADV":
            counts["adverbs"] += 1

    total_counts["nouns"] += counts["nouns"]
    total_counts["verbs"] += counts["verbs"]
    total_counts["adjectives"] += counts["adjectives"]
    total_counts["adverbs"] += counts["adverbs"]
    return f"nouns = {counts['nouns']}, verbs = {counts['verbs']}, adjectives = {counts['adjectives']}, adverbs = {counts['adverbs']}"


df_cleaned["POS_Counts"] = df_cleaned["Cleaned_text"].apply(pos)
```

```python
#2 Constituency and Dependency Parsing

import spacy
import benepar
import nltk
from nltk.tree import Tree
import pandas as pd

# Initialize spaCy model for processing text
nlp_model = spacy.load("en_core_web_sm")

# Download required NLTK and Benepar files
nltk.download('punkt')
benepar.download('benepar_en3')
nltk.download('punkt_tab')
benepar_parser = benepar.Parser("benepar_en3")

# Helper function to split long sentences into smaller chunks for parsing
def split_sentence_into_chunks(sentence, max_token_limit=512):
    doc = nlp_model(sentence)
    sentence_chunks = []
    current_chunk = []

    for token in doc:
        current_chunk.append(token.text)
        if len(current_chunk) > max_token_limit:
            sentence_chunks.append(" ".join(current_chunk[:-1]))  # Add the chunk without the last token
            current_chunk = [current_chunk[-1]]  # Start a new chunk with the last token

    if current_chunk:  # Add the last chunk
        sentence_chunks.append(" ".join(current_chunk))

    return sentence_chunks

# Function to generate and print constituency and dependency trees
def analyze_sentence_structure(text):
    doc = nlp_model(text)

    # Print dependency parse results
    print(f"\nDependency Parsing for the sentence: {text}")
    for word in doc:
        print(f"{word.text} --> {word.dep_} ({word.head.text})")

    # Print constituency parse trees
    for sentence in doc.sents:
        sentence_chunks = split_sentence_into_chunks(sentence.text)
        for chunk in sentence_chunks:
            print("\nConstituency Parse Tree for chunk:")
            try:
                tree = list(benepar_parser.parse_sents([chunk]))[0]
                tree.pretty_print()
            except ValueError as error:
                print(f"Error parsing chunk: {error}")

# Process only the first 2 rows as examples
sample_rows = df_cleaned['Cleaned_text'].head(2)  # Adjust if you want more rows

# Apply parsing function to these rows
for review in sample_rows:
    analyze_sentence_structure(review)
```

```python
#3 Named Entity Recognition
#Named Entity Recognition: Extract all the entities such as person names, organizations,
#locations, product names, and date from the clean texts, calculate the count of each entity.
import spacy

spacy_eng_model = spacy.load("en_core_web_sm")

def ner(features):
  ner_model = spacy_eng_model(features)
  entities_count = {
      "person_names": 0,
      "organizations": 0,
      "locations": 0,
      "product_names": 0,
      "dates": 0
  }

  for words in ner_model.ents:
    if words.label_ == "PERSON":
      entities_count["person_names"] += 1
    elif words.label_ == "ORG":
      entities_count["organizations"] += 1
    elif words.label_ == "LOC":
      entities_count["locations"] += 1
    elif words.label_ == "PRODUCT":
      entities_count["product_names"] += 1
    elif words.label_ == "DATE":
      entities_count["dates"] += 1

  return f"person_names = {entities_count['person_names']}, organizations = {entities_count['organizations']}, locations = {entities_count['locations']}, product_names = {entities_count['product_names']}, dates = {entities_count['dates']}"


df_cleaned["NER_Counts"] = df_cleaned["Cleaned_text"].apply(ner)
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Load the CSV file into a DataFrame
file_path = '/content/drive/MyDrive/github_marketplace_products.csv'
df2 = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df2.head())

# Ensure the 'Page Number' column exists
if 'Page Number' not in df2.columns:
    df2['Page Number'] = (df2.index // 20) + 1

# Define the preprocessing function
def preprocess_text(text):
    # Tokenization
    tokens = word_tokenize(text)
    # Converting to Lowercase
    tokens = [word.lower() for word in tokens]
    # Removing Stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    # Join tokens back into a single string
    cleaned_text = ' '.join(filtered_tokens)
    return cleaned_text

# Apply the preprocessing function to the text column
# Assuming the text column is named 'Description'
df2['cleaned_description'] = df2['Description'].apply(preprocess_text)

# Display the first few rows of the DataFrame with cleaned text and Page Number
print(df2[['Description', 'cleaned_description', 'Page Number']].head())

# Save the updated DataFrame back to the same CSV file
df2.to_csv(file_path, index=False)
print(f"Updated data saved to {file_path}")

```

```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from urllib.parse import urlparse

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load the CSV file into a DataFrame
file_path = '/content/drive/MyDrive/github_marketplace_products.csv'  # Replace with your CSV file path
df2 = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df2.head())

# Ensure the 'Page Number' column exists
if 'Page Number' not in df2.columns:
    df2['Page Number'] = (df2.index // 20) + 1

# Define the preprocessing function
def preprocess_text(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Remove unnecessary whitespace
    text = re.sub(r'\s+', ' ', text)
    # Tokenization
    tokens = word_tokenize(text)
    # Converting to Lowercase
    tokens = [word.lower() for word in tokens]
    # Removing Stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    filtered_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    # Join tokens back into a single string
    cleaned_text = ' '.join(filtered_tokens)
    return cleaned_text

# Apply the preprocessing function to the text column
df2['cleaned_description'] = df2['Description'].apply(preprocess_text)

# Data Quality Checks

# 1. Completeness: Check for missing values and handle them
required_columns = ['Product Name', 'Product URL', 'Description', 'Page Number']
df2.dropna(subset=required_columns, inplace=True)

# 2. Consistency: Ensure data is formatted correctly
# Check if 'Product Name' and 'Description' are strings
df2['Product Name'] = df2['Product Name'].astype(str)
df2['Description'] = df2['Description'].astype(str)

# Check if 'Product URL' is a valid URL
def is_valid_url(url):
    try:
        result = urlparse(url)
        return all([result.scheme, result.netloc])
    except ValueError:
        return False

# Filter out rows with invalid URLs
df2 = df2[df2['Product URL'].apply(is_valid_url)]

# 3. Accuracy: Remove duplicates
df2.drop_duplicates(inplace=True)

# Display the first few rows of the DataFrame with cleaned text and Page Number
print(df2[['Product Name', 'Product URL', 'Description', 'cleaned_description', 'Page Number']].head())

# Save the updated DataFrame back to the same CSV file
df2.to_csv(file_path, index=False)
print(f"Updated data saved to {file_path}")
```

```python
client = tweepy.Client(bearer_token=BEARER_TOKEN)

# 2. Improved Query Structure
query = "(#machinelearning OR #artificialintelligence) -is:retweet lang:en"

# 3. Proper Field Management
tweet_fields = ["id", "text", "author_id", "created_at"]
user_fields = ["username"]
expansions = ["author_id"]

tweets_data = []
users_cache = {}
next_token = None  # Crucial for pagination
retry_count = 0
max_retries = 3

try:
    while len(tweets_data) < 100 and retry_count < max_retries:
        try:
            # 4. API Call with Pagination
            response = client.search_recent_tweets(
                query=query,
                tweet_fields=tweet_fields,
                user_fields=user_fields,
                expansions=expansions,
                max_results=100,  # Max allowed per request
                next_token=next_token
            )

            # 5. Handle User Data First
            if response.includes and 'users' in response.includes:
                users_cache.update({u.id: u.username for u in response.includes['users']})

            # 6. Process Tweets
            if response.data:
                for tweet in response.data:
                    tweets_data.append({
                        'tweet_id': tweet.id,
                        'text': tweet.text,
                        'author_id': tweet.author_id,
                        'username': users_cache.get(tweet.author_id, 'N/A'),
                        'created_at': tweet.created_at
                    })

                print(f"Collected {len(tweets_data)} tweets")

                # 7. Pagination Handling
                next_token = response.meta.get('next_token', None)

                if not next_token or len(tweets_data) >= 100:
                    break

                # 8. Proper Rate Limiting
                time.sleep(5)  # Essential wait between requests

        except tweepy.TooManyRequests:
            print("Rate limit exceeded. Waiting 15 minutes...")
            time.sleep(900)  # 15 minute wait
            retry_count += 1

        except tweepy.TweepyException as e:
            print(f"API Error: {e}")
            retry_count += 1
            time.sleep(60)

except Exception as e:
    print(f"Critical error: {e}")

# 9. Final Data Handling
if tweets_data:
    df = pd.DataFrame(tweets_data)
    print("\nSample data:")
    print(df.head())

    df.to_csv('tweets.csv', index=False)
    print(f"\nSuccessfully saved {len(df)} tweets to tweets.csv")
else:
    print("No tweets were collected")
```

# Part 2 Perform data cleaning procedures

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def clean_tweet_text(text):
    # Remove URLs
    text = re.sub(r"http\S+|www\S+|https\S+", "", text, flags=re.MULTILINE)
    # Remove mentions (@username)
    text = re.sub(r"@\w+", "", text)
    # Remove special characters and numbers (keep only letters and spaces)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    # Convert to lowercase
    text = text.lower()
    # Remove extra spaces
    text = " ".join(text.split())
    return text

df["cleaned_tweets"] = df["text"].apply(clean_tweet_text)
```
